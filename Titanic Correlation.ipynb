```python
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd

# Load your DataFrame from a CSV file
df = pd.read_csv('train.csv')

# Check data types to identify non-numeric columns
print(df.dtypes)

# Drop rows with missing values
df = df.dropna()

# Calculate the correlation between 'Survived' and 'Age'
correlation_age_survived = df['Survived'].corr(df['Age'])
print(f"Correlation between 'Survived' and 'Age': {correlation_age_survived}")

# Calculate the correlation matrix
correlation_matrix = df.corr(numeric_only = True)

# Display the correlation matrix using a heatmap
sn.heatmap(correlation_matrix, annot=True)
plt.show()
```
